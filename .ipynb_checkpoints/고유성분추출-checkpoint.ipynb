```python
import pandas as pd
import re

df = pd.read_excel('통합성분.xlsx')

all_ingredients = set()
for val in df['전성분'].dropna():
    val = str(val).strip().strip('"')
    
    # @ 또는 , 로 분리
    ingres = re.split(r'[@,]', val)
    ingres = [i.strip() for i in ingres if i.strip()]
    all_ingredients.update(ingres)

# 빈 문자열 제거
all_ingredients.discard('')

# 고유 성분 사전 저장
ingre_dict_df = pd.DataFrame({
    'ingre_name': sorted(list(all_ingredients)),
    'recommended_skin': None,
    'sensitive_skin': None,
    'allergy_check': None
})

print(f'고유 성분 수: {len(ingre_dict_df)}개')
ingre_dict_df.to_csv('ingredient_dict.csv', index=False, encoding='utf-8-sig')
print('✅ ingredient_dict.csv 저장 완료!')
ingre_dict_df.head(10)
```
