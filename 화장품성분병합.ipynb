```python
import pandas as pd

# 두 파일 불러오기
reviews_df = pd.read_csv('cleaned_reviews.csv', encoding='utf-8-sig')
ingredients_df = pd.read_csv('성분표.csv', encoding='euc-kr')

# 제품명 기준 LEFT JOIN
merged_df = reviews_df.merge(
    ingredients_df[['제품명', '전성분']],
    on='제품명',
    how='left'
)

print(f'전체 리뷰: {len(merged_df)}')
print(f'성분 매칭 성공: {merged_df["전성분"].notna().sum()}')
print(f'성분 없음: {merged_df["전성분"].isna().sum()}')

# 결과 저장
merged_df.to_csv('reviews_ingredients.csv', index=False, encoding='utf-8-sig')
print('✅ reviews_with_ingredients.csv 저장 완료!')
merged_df.head(3)
```
